```
import scipy.integrate as spi
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# n tendra la funcion de asignar valores de 0 a 10 
# para que se generen valores de la integral.
n = np.arange(0,10,0.1)

# Escribi el metodo de trapecio dos veces para resolver S y C
def trapesicion(a,b,N=50):
    x = np.linspace(a,b,N+1)
    y = np.sin(x**2)
    y_right = y[1:] 
    y_left = y[:-1] 
    dx = (b - a)/N
    T = (dx/2) * np.sum(y_right + y_left)
    return T
  
def trapesicion2(a,b,N=50):
    x = np.linspace(a,b,N+1)
    y = np.cos(x**2)
    y_right = y[1:] 
    y_left = y[:-1] 
    dx = (b - a)/N
    T = (dx/2) * np.sum(y_right + y_left)
    return T

  
# Vectorice las funciona para asi despues graficarlar, dado que estas 
# general un arreglo de una dimension.

Vp = np.vectorize(trapesicion)  
Vp2 = np.vectorize(trapesicion2)  

Fv = Vp(0,n,139)
Fv2 = Vp2(0,n,139)


# Graficacion, solo se hizo S en funcion de C.
plt.plot(-Fv2,-Fv)
plt.plot(Fv2,Fv)
plt.ylabel('Espiral de Euler')
plt.xlabel('x')
plt.title('Grafica por Matthew Ruvalcaba')
plt.grid()
plt.show()
```
